```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Création du graphe
G = nx.DiGraph()

# Ajout des nœuds (concepts)
concepts = ["chien", "animal", "viande", "nourriture", "chat", "être vivant", "respirer"]
G.add_nodes_from(concepts)

# Ajout des arêtes (relations)
relations = [
    ("chien", "animal", "est un"),
    ("chien", "viande", "mange"),
    ("viande", "nourriture", "est un"),
    ("chat", "animal", "est un"),
    ("animal", "être vivant", "est un"),
    ("être vivant", "respirer", "a besoin de")
]
```

```python
# Ajout des arêtes avec leurs étiquettes
for source, target, relation in relations:
    G.add_edge(source, target, label=relation)

# Dessin du graphe avec NetworkX et Matplotlib
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)  # Génération d'une disposition pour le graphe

# Dessin des nœuds et des arêtes
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=10, font_weight="bold", arrows=True)

# Dessin des étiquettes des arêtes (relations)
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red', font_size=9)

plt.title("Réseau Sémantique des Concepts et Relations")
plt.show()
```

```python
# Fonction pour trouver les relations associées à un concept donné
def find_relations(concept):
    relations = list(G.edges(concept, data=True))
    print(f"Relations associées au concept '{concept}':")
    for src, tgt, data in relations:
        print(f"{src} --{data['label']}--> {tgt}")
```

```python
# Fonction pour afficher la relation entre deux concepts spécifiques
def find_direct_relation(concept1, concept2):
    if G.has_edge(concept1, concept2):
        label = G[concept1][concept2]['label']
        print(f"Relation entre '{concept1}' et '{concept2}': {concept1} --{label}--> {concept2}")
    else:
        print(f"Aucune relation directe entre '{concept1}' et '{concept2}'")
```

```python
# Exemple de recherche de relations
find_relations("chien")
find_direct_relation("chien", "animal")
```
